```
from bs4 import BeautifulSoup
import re
```

```
html = requests.get("https://search-merchant.xn--42caj4e6bk1f5b1j.com/shop/c1faab20fc2af4dd38148ac347e29504")
bs = BeautifulSoup(html.content, "html.parser")
print(bs.prettify())
```

```
for index, row in df.iterrows():
  print(row['url'])
  html = requests.get(row['url'])
  bs = BeautifulSoup(html.content, "html.parser")
  result_a_by_class = bs.find('a', attrs={'class': 'break-word'})

  if result_a_by_class:
    txt = str(result_a_by_class.get_attribute_list('href')[0])
    x = re.split("/", txt)
    lat, lon = re.split(",",x[-2])
    print(f'lat:{lat} ,lon:{lon}')
    df.loc[index,'lat'] = lat
    df.loc[index,'long'] = lon

  
```
